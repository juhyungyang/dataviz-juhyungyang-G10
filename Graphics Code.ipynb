```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
from matplotlib import rcParams
```

```python
#The 10 countries that lie on the equator are
#Ecuador, Colombia, Brazil, Gabon, Sao Tome & Principe, 
#Democratic Republic of the Congo, Uganda, Kenya, Somalia, and Indonesia.

countries_on_equator = ['Ecuador', 'Colombia', 'Brazil', 'Gabon', 'Sao Tome & Principe',
                       'D.R. of the Congo', 'Uganda', 'Kenya', 'Somalia','Indonesia']

#The average temperature of those countries (1991 - 2020) in (C)
ecu_temp = 21.65
col_temp = 24.79
bra_temp = 25.51
gab_temp = 25.36
sao_temp = 23.20
DRC_temp = 24.36
uga_temp = 23.34
ken_temp = 25.09
som_temp = 27.09
ind_temp = 26.14

avg_temp_equator = [ecu_temp, col_temp, bra_temp, gab_temp, sao_temp,
                   DRC_temp, uga_temp, ken_temp, som_temp, ind_temp]

#The 10 countries furthest from the equator are 
#Greenland (Denmark), Canada, Russia, Norway, U.S.A, 
#Finland, Sweden, Iceland, U.K., and Estonia

countries_furthest = ['Greenland', 'Canada', 'Russia', 'Norway', 'U.S.A.',
                     'Finland', 'Sweden', 'Iceland', 'U.K.', 'Estonia']

#The average tmeperature of those countires (1991 - 2020) in (C)
gre_temp = -17.18
can_temp = -6.36
rus_temp = -5.16
nor_temp = -1.78
USA_temp = 7.63
fin_temp = 2.52
swe_temp = 2.62
ice_temp = 2.31
UK_temp = 9.07
est_temp = 6.31

avg_temp_furthest = [gre_temp, can_temp, rus_temp, nor_temp, USA_temp,
                    fin_temp, swe_temp, ice_temp, UK_temp, est_temp]
```

# Countries on Equator vs. Average Temperature

```python
#Creating dataframe of Countries on Equator vs. Their Average Temperature
equator_df = pd.DataFrame(avg_temp_equator,
                  index= countries_on_equator, 
                  columns = pd.Index(['Temperature in °C'])) 
equator_df
```

```python
#Countries on Equator vs. Their Average Temperature (BAR Plot 1)
equator = equator_df.plot.barh(legend=False, color='r')

equator.set_xlim(-20,28)
equator.set_xlabel('Avg. Temp. (°C)')
equator.set_ylabel('Countries on Equator')
equator.set_title('Countries on Equator vs. Avg. Temp.')
```

# Countries Furthest from Equator vs. Average Temperature 

```python
#Creating dataframe of Countries Furthest from Equator vs. Their Average Temperature
furthest_df = pd.DataFrame(avg_temp_furthest,
                  index= countries_furthest, 
                  columns = pd.Index(['Temperature in °C'])) 
furthest_df
```

```python
#Countries Furthest from Equator vs. Their Average Temperature (BAR Plot 2)
furthest = furthest_df.plot.barh(legend=False)

furthest.set_xlabel('Avg. Temp. (°C)')
furthest.set_ylabel('Countries Furthest from Equator')
furthest.set_title('Countries Furthest from Equator vs. Avg. Temp.')
```

# Countries VS. Total Covid-19 Cases VS. Average Temperature

```python
#Comparing all three variables
#Country VS. Total Covid-19 Cases VS. Average Temperature
combined_df = pd.read_csv('countries_data.csv')
combined_df = combined_df.sort_values(by='Total Cases (In Thousands)', ascending=False)

combined_df
```

```python
#Plotting with pandas 
ax = combined_df.plot.barh(x='Total Cases (In Thousands)', y='Avg. Temp.', legend=False)

ax.set_xlabel('Avg. Temp. (°C)')
ax.set_title('Avg. Temp. vs. Total Covid Cases per Country ')
```

```python
#Plotting with seaborn (barplot)
final = sns.barplot(y='Total Cases (In Thousands)', x='Avg. Temp.', data= combined_df)

final.set_title('Avg. Temp. vs. Total Covid-19 Cases')
sns.set_theme(style = 'darkgrid')
sns.set_palette("Paired")
rcParams['figure.figsize'] = 12,6
```

```python
#Plotting with seaborn (lineplot)
final_line = sns.lineplot(y='Total Cases (In Thousands)', x='Avg. Temp.', data= combined_df, marker="o", color='red')

final_line.set_title('Avg. Temp. vs. Total Covid-19 Cases')
sns.set_style('darkgrid')
rcParams['figure.figsize'] = 12,8
```

```python
#Plotting both bar & line plot
fig, ax1 = plt.subplots(figsize=(12,6))

sns.lineplot(data = combined_df, x='Avg. Temp.', y='Total Cases (In Thousands)', marker='o', ax=ax1, color='red')
ax2 = ax1.twinx()

sns.barplot(data = combined_df, x='Avg. Temp.', y='Total Cases (In Thousands)', ax=ax2)
sns.set_style('white')
```
